```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# For QQ plots
import statsmodels.api as sm
```

```python
path = ''
fileName = ''

df = pd.read_csv(path+fileName)
```

```python
def qqplot(df):
    # Create Q-Q plot with 45-degree line
    fig = sm.qqplot(df, line='45')
    plt.show()
    

qqplot(df['Ed'])    
```
