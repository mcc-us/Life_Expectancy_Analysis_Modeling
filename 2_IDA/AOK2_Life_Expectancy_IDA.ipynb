Purpose: This notebook investigates *missing values, renames features, and imputes missing values by column mean*.

Input: `Life_Expectancy_Data.csv`

Output: `Clean_LE_Data_w_Means_1.csv`

```python
# Common Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Load, rename columns, dataset

```python
filename = '/home/mcc/Desktop/Life_Expectancy_Analysis_Modeling/data/raw/Life_Expectancy_Data.csv'

column_names = ['Country','Year','Status','LifeExpectancy','AdultMort',
                'InfD','EtOH','PercExpen','HepB','Measles',
                'BMI','lt5yD','Polio','TotalExpen','DTP','HIV',
                'GDP','Population','Thin1_19y','Thin5_9y','Income',
                'Education']

df = pd.read_csv(filename, names=column_names, header=0)

df.info()
```

## Check number of null values

```python
print('='*38)
print('Column              No. missing values')
print('='*38)
print(df.isnull().sum())
```

### NOTE 1: 
- The **Dependent feature** 'LifeExpectancy' has 10 missing values. 
 
- Q. What exactly is missing from LifeExpectancy column?

```python
df.loc[df['LifeExpectancy'].isnull()]
```

```python
# Produces one entry
df.loc[df['Country'] == 'Cook Islands']
```

### Note 2:

- Further investigation (not shown here) finds that all ten missing values are from countries with only one row (year=2013). 

- 'LifeExpectancy' is a Dependent variable. I will delete(drop) those 10 observations with NO labels rather than impute them.

- 'LifeExpectancy' appears to be **Missing Completely at Random(MCAR)**. The advantage of **MCAR** is that the analysis is unbiased.

```python
# 10 country rows deleted
df.dropna(subset=['LifeExpectancy'], inplace=True)
```

```python
print('\nDataframe shape:', df.shape)

df.isnull().sum()
```

### NOTE 2: 
- Three features with the highest number of mising vlaues (out of 2928) are:

|  | Feature | Number Missing | % Missing |
|--|:--------|---------------:|----------:|
|1 | Population | 644 | 22.0% |
|2 | HepB | 553 |  18.9% |
|3 | GDP | 448 | 15.3% |

- ['Population', 'HepB', 'GDP'] are missing greater than 15%.

- Because these three features are missing more than 15%, they will be deleted from analysis.

```python
df.drop(['Population', 'HepB', 'GDP'], axis=1, inplace=True)
```

## Imputation using column means

```python
df['InfD'].fillna(np.mean(df.InfD), inplace=True)
df['EtOH'].fillna(np.mean(df.EtOH), inplace=True)
df['PercExpen'].fillna(np.mean(df.PercExpen), inplace=True)
df['Measles'].fillna(np.mean(df.Measles), inplace=True)
df['BMI'].fillna(np.mean(df.BMI), inplace=True)
df['Polio'].fillna(np.mean(df.Polio), inplace=True)
df['TotalExpen'].fillna(np.mean(df.TotalExpen), inplace=True)
df['DTP'].fillna(np.mean(df.DTP), inplace=True)
df['Thin1_19y'].fillna(np.mean(df.Thin1_19y), inplace=True)
df['Thin5_9y'].fillna(np.mean(df.Thin5_9y), inplace=True)
df['Income'].fillna(np.mean(df.Income), inplace=True)
df['Education'].fillna(np.mean(df.Education), inplace=True)

# Convert Dev status to a binary variable, where Developing Nations = 0, Developed Nation status = 1.
df['Status'] = df['Status'].apply(lambda x: 0 if x == 'Developing' else 1).astype('int8')
```

## Re-Check null data points
